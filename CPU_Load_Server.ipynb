```python
# ðŸ“Š Chicha's CPU Load Graph for Nani ðŸš€

import pandas as pd
import matplotlib.pyplot as plt

# Upload your CSV (manually, once)
from google.colab import files
uploaded = files.upload()

# Load the CSV into a DataFrame
df = pd.read_csv("report_data.csv")

# Convert Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Plot CPU Load over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["CPU_Load"], marker='o', color='blue', label="CPU Load (%)")

plt.title("CPU Load Over Time")
plt.xlabel("Date")
plt.ylabel("CPU Load (%)")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
